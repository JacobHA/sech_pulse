```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Draw an energy level diagram for a two-level atom driven by a laser.
# The energy levels are labeled as |g>, |e>
# The transition frequencies are labeled as omega_21. Draw an incoming laser as a red squiggly arrow.

energy_g = 0
energy_e = 1.0
delta = 0.15

plt.ylim(-0.3,1.3)
plt.xlim(-1,1)
# turn off the axis
plt.axis('off')

# Draw a thick line for each level.
width=1
gap=-0.25
x10 = 0
x11 = width
xe0 = 0
xe1 = xe0 + width

center1 = (x10+x11)/2
centere = (xe0+xe1)/2

plt.hlines(energy_g, x10, x11, 'k', linewidth=5)
plt.hlines(energy_e, xe0, xe1, 'k', linewidth=5)

# matplotlib color code for grey:
# https://matplotlib.org/2.0.2/api/colors_api.html
color = '#808080'
plt.hlines(energy_e - delta, xe0, xe1, color, linewidth=5)
# Draw a small double-headed arrow between e and e- delta:
plt.arrow(xe1+0.1, energy_e, 0, -delta+0.02, head_width=0.05, head_length=0.03, fc='k', ec='k')
plt.arrow(xe1+0.1, energy_e-delta, 0, delta-0.02, head_width=0.05, head_length=0.03, fc='k', ec='k')
# label this arrow as delta(t):
plt.text(xe1-0.7, energy_e-delta/2, r'$\Delta(t)$', fontsize=16, horizontalalignment='center', verticalalignment='center')

# label the energy levels
plt.text(center1, energy_g-0.07, r'$\vert 0 \rangle$', fontsize=16, horizontalalignment='center', verticalalignment='center')
plt.text(centere, energy_e+0.07, r'$\vert 1 \rangle$', fontsize=16, horizontalalignment='center', verticalalignment='center')
# Draw a thin arrow for each transition.
# transition from 1->e
p1 = patches.FancyArrowPatch((center1, energy_g), (centere, energy_e), arrowstyle='<|-|>', mutation_scale=20, color='r')
plt.gca().add_patch(p1)
# label the arrow with omega:
plt.text(1.15*center1, (energy_g+energy_e)/2, r'$\Omega(t)$', fontsize=16, horizontalalignment='center', verticalalignment='center')
# draw a sine wave for the laser
t = np.linspace(-0.0, 0.45, 100)
plt.plot(t, 0.1*np.sin(30*t+0.4)+0.5, 'r', linewidth=3)


plt.xlim(-0.,0.7)
save = True
if save:
    plt.savefig('tls.png', dpi=300)
```

```python
# Draw an energy level diagram for the Lambda-type system.
# The energy levels are labeled as |1>, |2>, and |e>.
# The transition frequencies are labeled as omega_21 and omega_e1.

energy_1 = 0
energy_2 = 0.2
energy_e = 1.0
delta = 0.15

plt.ylim(-0.3,1.3)
plt.xlim(0,3)
# turn off the axis
plt.axis('off')

# Draw a thick line for each level.
width=1
gap=-0.25
x10 = 0
x11 = width
xe0 = x11 + gap
xe1 = xe0 + width
x20 = xe1 + gap
x21 = x20 + width
center1 = (x10+x11)/2
center2 = (x20+x21)/2
centere = (xe0+xe1)/2

plt.hlines(energy_1, x10, x11, 'k', linewidth=5)
plt.hlines(energy_e, xe0, xe1, 'k', linewidth=5)
plt.hlines(energy_2, x20, x21, 'k', linewidth=5)

# matplotlib color code for grey:
# https://matplotlib.org/2.0.2/api/colors_api.html
color = '#808080'
plt.hlines(energy_e - delta, xe0, xe1, color, linewidth=5)
# Draw a small double-headed arrow between e and e- delta:
plt.arrow(xe1+0.1, energy_e, 0, -delta+0.02, head_width=0.05, head_length=0.03, fc='k', ec='k')
plt.arrow(xe1+0.1, energy_e-delta, 0, delta-0.02, head_width=0.05, head_length=0.03, fc='k', ec='k')
# label this arrow as delta(t):
plt.text(xe1+0.3, energy_e-delta/2, r'$\Delta(t)$', fontsize=16, horizontalalignment='center', verticalalignment='center')

# label the energy levels
plt.text(center1, energy_1-0.07, r'$\vert 0 \rangle$', fontsize=16, horizontalalignment='center', verticalalignment='center')
plt.text(center2, energy_2-0.07, r'$\vert 1 \rangle$', fontsize=16, horizontalalignment='center', verticalalignment='center')
plt.text(centere, energy_e+0.07, r'$\vert e \rangle$', fontsize=16, horizontalalignment='center', verticalalignment='center')
# Draw a thin arrow for each transition.
# transition from 1->e
p1 = patches.FancyArrowPatch((center1, energy_1), (centere, energy_e), arrowstyle='<|-|>', mutation_scale=20, color='r')
plt.gca().add_patch(p1)
# label the arrow with omega:
plt.text(1.2*center1, (energy_1+energy_e)/2, r'$\Omega(t)\omega_0$', fontsize=16, horizontalalignment='center', verticalalignment='center')
# do the same for 2->e
p2 = patches.FancyArrowPatch((center2, energy_2), (centere, energy_e), arrowstyle='<|-|>', mutation_scale=20, color='r')
plt.gca().add_patch(p2)
plt.text(1.55*centere, (energy_2+energy_e)/2, r'$\Omega(t)\omega_1$', fontsize=16, horizontalalignment='center', verticalalignment='center')

save = True
if save:
    plt.savefig('lambda.png', dpi=300)
```

```python
# Plot the step function Theta:
def Theta(t):
    if t < 0:
        return 0
    else:
        return 1
    
t = np.linspace(-1, 1, 1000)
y = [Theta(t_i) for t_i in t]
plt.plot(t, y, 'k', linewidth=2)
plt.yticks([0,1], [r'$0$', r'$1$'], fontsize=16)
plt.xticks([-1,0,1], [-1,0,1], fontsize=16)
plt.savefig('heaviside.png', dpi=300)
```

```python
# similarly, plot relu, sigmoid and tanh:
def relu(t):
    if t < 0:
        return 0
    else:
        return t

def sigmoid(t):
    return 1/(1+np.exp(-t))

def tanh(t):
    return np.tanh(t)

t = np.linspace(-3, 3, 1000)
y = [relu(t_i) for t_i in t]
plt.plot(t, y, 'k', linewidth=2, label=r'ReLU')
z = [sigmoid(t_i) for t_i in t]
plt.plot(t, z, 'r', linewidth=2, label=r'Sigmoid')
w = [tanh(t_i) for t_i in t]
plt.plot(t, w, 'b', linewidth=2, label=r'Tanh')
plt.ylim(-2,2)
plt.yticks([-1,0,1], [-1,0,1], fontsize=16)
plt.xticks([-3,-2,-1,0,1,2,3], [-3,-2,-1,0,1,2,3], fontsize=16)
plt.legend(fontsize=16)
# draw dashed lines at y=0 and y=1 y=-1::
plt.plot([-3,3], [0,0], 'k--', alpha=0.5)
plt.plot([-3,3], [1,1], 'k--', alpha=0.5)
plt.plot([-3,3], [-1,-1], 'k--', alpha=0.5)
# draw an arrowhead at top of relu
x_arrow = 1.8
y_arrow = relu(x_arrow)
# tilt the arrowhead to point along y=x:
# plt.arrow(x_arrow, y_arrow, 0.1, 0.1, head_width=0.13, head_length=0.13, fc='k', ec='k')

plt.savefig('activation_functions.png', dpi=300)
```
